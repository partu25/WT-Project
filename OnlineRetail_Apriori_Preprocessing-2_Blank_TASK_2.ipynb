# Step 1: Load Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('OnlineRetail.csv')
df
```

# Step 2: Drop Missing Invoices or Descriptions

```python
df.dropna(subset=['InvoiceNo','Description'])
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

# Step 4: Keep Only Positive Quantities

```python
df['Quantity'].value_counts()
```

```python
df = df[df['Quantity']>0]
df
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df['Description'] = df['Description'].str.lower().str.replace(' ','')
```

```python
df
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
df = df[df['Country'] == 'United Kingdom'] 
df
```

```python
df.columns
```

# Step 7: Remove Duplicates

```python
df.drop_duplicates()
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

# Step 9 : Check for Top 10 Selling Products

```python
df.groupby('Description')['Quantity'].sum().sort_values(ascending = False).head(10)
```

# Step 10 : Identify Most Active Customers

```python
df.groupby('CustomerID')['InvoiceNo'].nunique().sort_values(ascending = False).head(10)
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['Month'] = df['InvoiceDate'].dt.month
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
popular_months = df['Month'].value_counts().sort_index()
popular_months
```

```python
import matplotlib.pyplot as plt
popular_months.plot(kind = 'bar')
plt.show()
```

```python
popular_hour = df['Hour'].value_counts().sort_index()
popular_hour
```

```python
popular_hour.plot(kind= 'bar')
plt.show()
```

```python
df = df[df['Quantity']>0]
```

```python
df
```

```python
df.dropna(subset= ['InvoiceNo','Description'],inplace= False)
```

```python
basket = df.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0)
```

```python
basket
```

```python
basket_full =basket.map(lambda x : 1 if x>0 else 0 )
```
